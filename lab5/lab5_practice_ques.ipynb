```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plot tan(x), cot(x), sec(x) and cosec(x) for the values of x= [-pi,-pi/4, -pi/2, 0, pi/4, pi/2, pi].

X = [-np.pi, -np.pi/4, -np.pi/2, 0, np.pi/4, np.pi/2, np.pi]
Tan = np.arctan(X)
Cot = np.arctan(X)
Sec = np.arccos(X)
Cosec = np.arcsin(X)

plot1 = plt.figure("Tan")
plt.plot(X, Tan, color = "blue")
plot2 = plt.figure("Cot")
plt.plot(X, Cot, color = "red")
plot3 = plt.figure("Sec")
plt.plot(X, Sec, color = "green")
plot4 = plt.figure("Cosec")
plt.plot(X, Cosec, color = "yellow")
```

```python
# Represent the following table using bar chart:
# Method Result1 Result2
# A 2 3
# B 5 2
# C 8 5
# D 5 7


X = np.array(['A', 'B', 'C', 'D'])
R1 = np.array([2, 5, 8, 5])
R2 = np.array([3, 2, 5, 7])
X1 = np.arange(len(R1))
X2 = [x + 0.25 for x in X1]
plt.bar(X1, R1, facecolor='#9999ff', edgecolor='white', width = 0.25)
plt.bar(X2, R2, facecolor='#ff9999', edgecolor='white', width = 0.25)
plt.xticks([x+ 0.1 for x in X1], X)
plt.show()
```
